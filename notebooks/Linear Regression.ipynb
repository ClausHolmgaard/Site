```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Creating some data

```python
slope = 2
rand_factor = 15

x = np.arange(0, 100)
y = np.random.normal(size=100) * rand_factor + x * (slope - 1)
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(x, y)
plt.show()
```
